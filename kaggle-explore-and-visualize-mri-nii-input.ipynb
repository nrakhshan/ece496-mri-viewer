```python
import importlib, subprocess, sys
for pkg in ['nibabel', 'skimage', 'plotly']:
    if importlib.util.find_spec(pkg) is None:
        subprocess.run([sys.executable, '-m', 'pip', 'install', '-q',
                        'scikit-image' if pkg == 'skimage' else pkg])

import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt


NII_PATH = '/kaggle/input/datasets/awsaf49/brats20-dataset-training-validation/BraTS2020_TrainingData/MICCAI_BraTS2020_TrainingData/BraTS20_Training_297/BraTS20_Training_297_flair.nii'


img = nib.load(NII_PATH)
data = img.get_fdata()

# fMRI/4D files have a time axis — take the first volume
if data.ndim == 4:
    data = data[..., 0]

zooms = img.header.get_zooms()[:3]   # voxel size in mm
print('Shape:', data.shape)
print('Voxel spacing (mm):', zooms)
print('Intensity range:', float(data.min()), 'TO', float(data.max()))


plt.figure(figsize=(7, 3))
plt.hist(data.ravel(), bins=100)
plt.yscale('log'); plt.xlabel('Intensity'); plt.ylabel('Voxel count (log)')
plt.title('Intensity histogram'); plt.show()



def show_slices(vol, spacing=(1, 1, 1)):
    x, y, z = np.array(vol.shape) // 2
    sx, sy, sz = spacing
    planes = [
        (vol[x, :, :].T, 'Sagittal', sz / sy),
        (vol[:, y, :].T, 'Coronal',  sz / sx),
        (vol[:, :, z].T, 'Axial',    sy / sx),
    ]
    fig, axes = plt.subplots(1, 3, figsize=(13, 5))
    for ax, (sl, title, aspect) in zip(axes, planes):
        ax.imshow(sl, cmap='gray', origin='lower', aspect=aspect)
        ax.set_title(title); ax.axis('off')
    plt.tight_layout(); plt.show()

show_slices(data, zooms)

# axis: direction to slice along 
# n = num of slices 
# columns
# data

def montage(vol, n=16, axis=2, cols=4):\
    #Slices are 0 indexed
    # [0, l, k ...] equally spaced numbers in the array (slice number)
    idx = np.linspace(0, vol.shape[axis] - 1, n).astype(int)
    rows = int(np.ceil(n / cols))
    fig, axes = plt.subplots(rows, cols, figsize=(3 * cols, 3 * rows))
    #axes [a11, a12 ....;a21, a22 ..]
    #axes.ravel() converts the matrix above to 1D
    # Each element will map to particular index slice
    for ax, i in zip(axes.ravel(), idx):
        ax.imshow(np.take(vol, i, axis=axis).T, cmap='gray', origin='lower')
        # np,take here is equivalent to vol[:,:,i]
        ax.set_title(f'{i}', fontsize=8); ax.axis('off')
    for ax in axes.ravel()[n:]:
        ax.axis('off')
    plt.tight_layout(); plt.show()

montage(data, n=16, axis=2)
#slice the brain horizontally 16 times from top to bottom and lay them out in a 4x4 grid.

```
